# Introdução do Google Colab

- O **Google Colab** é uma solução do Google para rodar no sistema de computação em nuvem do Google, notebooks em Python.
- Notebooks são maneiras interativas de executar código.
- Em um notebook, a cada bloco de código executado, o resultado é exibido na tela.
- O usuário controla o momento e a order que os blocos (**células**) serão executados.



### Executando código

O exemplo abaixo cria uma variável (número), soma 2 a esse número e mostra na o resultado.

Para executar o código bastar clicar no botão de "**play**" ao ladoda célula.

```
variavel = 5
variavel += 2
variavel
a =3
```

Alternativamente, para executar uma célula basta pressionar `Shift+Enter`.

```
variavel = 9
variavel += 3
variavel
```

### A ordem em que o código é executado, importa.

Tentar executar o código abaixo na ordem como está apresentada vai resultar em um erro.



```
b = 4
a = b * 3
a
```

```
b = 4
```

### Gráficos, imagens, tabelas...

- O grande poder de um  notebook é a possibilidade de produzir gráficos, imagens, tabelas ou outras tipos suportados, e  visualiza-los, ou editá-los, interativamente.

```
# Um gráfico

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y);
```

```
# Uma imagem
import matplotlib.cbook as cbook


with cbook.get_sample_data('grace_hopper.png') as image_file:
    image = plt.imread(image_file)

plt.imshow(image);
```
