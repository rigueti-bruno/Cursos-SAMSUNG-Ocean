Gráfico Simples com Matplolib

```python
# Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir os Dados
x = [1,2,3,4,5]
y = [4,10,11,7,5]
```

```python
#Plotar o Gráfico
plt.plot(x,y,'p--r')

# x - dados do eixo x
# y - dados do eixo y
# 'p--r' - string de formatação:
    # tipo de marcador: p (pentagono)
    # tipo de linha: -- (tracejada)
    # cor da linha: r (vermelho)
```

Gráficos de Camadas

```python
fig = plt.figure() # Cria a figura que receberá os gráficos
ax = plt.axes() # Cria os eixos de referência para os gráficos
```

```python
# dados do Eixo x:
x = np.array([1,2,3,4,5])
```

```python

fig = plt.figure() # Cria a figura que receberá os gráficos
ax = plt.axes() # Cria os eixos de referência para os gráficos

# Gráfico 1:
ax.plot(x, x*3, label='primeiro grafico') # label indica a referencia do grafico

# Gráfico 2:
ax.plot(x, x+2, label='segundo grafico')

# Inclui uma Legenda no gráfico:
ax.legend()

# Inclui um Grid no gráfico:
ax.grid(c='g', alpha=0.4, linestyle = '--')
# Onde:
    # c - cor das linhas (g)
    # alpha - transparência das linhas (0.4)
    # linestyle - tipo das linhas (--)

# Inclui um título para o Gráfico:
ax.set_title("título")

# Inclui titulos para os eixos do grafico:
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')

# Define a origem dos eixos para 0:
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
```

Gráficos Interativos com Plotly

```python
# Bibliotecas
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# Obtendo os Dados:
cotacao_bitcoin = pd.read_csv('https://github.com/ftorresd/OCEAN-datasets/raw/main/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-09-14.csv')
```

```python
# Dados estatísticos do dataset
cotacao_bitcoin.describe()
```

```python
# Criar uma coluna com as datas
cotacao_bitcoin['date'] = pd.to_datetime(cotacao_bitcoin['Timestamp'], unit='s')
cotacao_bitcoin
```

```python
# Agrupar os dados para facilitar a geração do gráfico
buffer = cotacao_bitcoin.groupby((cotacao_bitcoin.index // (60*24))).first()
```

```python
# Geração do Gráfico Interativo com Plotly
open = go.Scatter(x = buffer['date'], y = buffer['Open'], mode = 'lines', name = 'Abertura')

# Plotar o Gráfico em uma Figura
fig = go.Figure(open)

# Imprimir a figura
fig.show()
```
